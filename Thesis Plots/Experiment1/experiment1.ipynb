```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
before_1 = 8054
before_2 = 7779
after_1 = 6948
after_2 = 6866
```

```python
fig, ax = plt.subplots()
bottom = np.zeros(2)
width = 0.5

bar_names = ("Unfiltered", "Filtered")

weight_counts = {
    "Gardermoen-Flesland": np.array([before_1, after_1]),
    "Flesland-Gardermoen": np.array([before_2, after_2]),
}

for boolean, weight_count in weight_counts.items():
    p = ax.bar(bar_names, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

ax.legend(loc="upper right")
fig.savefig(f"Sample_counts.svg", format="svg", dpi=1000, bbox_inches='tight')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
p = ax.bar(["Gardermoen-Flesland", "Flesland-Gardermoen"], [before_1, before_2], edgecolor="black", bottom=1, log=True)
p.set_label("Before")

p = ax.bar(["Gardermoen-Flesland", "Flesland-Gardermoen"], [after_1, after_2], edgecolor="black", bottom=1, log=True)
p.set_label("After")

ax.legend(loc='lower right')
ax.set_ylabel("Frequency (log 10)")
plt.savefig("Sample_counts.svg", format="svg", bbox_inches="tight")
```
